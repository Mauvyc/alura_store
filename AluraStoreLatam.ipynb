### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
'''En este primer análisis, debes calcular el ingreso total de cada tienda. 
Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.'''

ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()


plt.bar(['Tienda 1','Tienda 2', 'Tienda 3', 'Tienda 4'],[ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4], 
        color=['#11608f', '#9d63b1', '#ff607b', '#ffa600'], 
        edgecolor='black'), plt.title('Grafico 1: Ingresos totales de las tiendas.')
plt.ticklabel_format(style='plain', axis='y')  # Para el eje Y

plt.show()
```

# 2. Ventas por categoría

```python

'''En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. 
La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.'''
vp1 = tienda.groupby('Categoría del Producto').size()
vp2 = tienda2.groupby('Categoría del Producto').size()
vp3 = tienda3.groupby('Categoría del Producto').size()
vp4 = tienda4.groupby('Categoría del Producto').size()

# Crear subplots: 2 filas x 2 columnas
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Grafico 2: Categorías de productos más y menos vendidas', fontsize=16)

# Gráfico para Tienda 1
axs[0, 0].pie(vp1, labels=vp1.index, colors=['#11608f','#4f65a7', '#8a64b1', '#c15eaa', '#ed5b92', '#ff666e', '#ff8244', '#ffa600'], 
              autopct='%1.1f%%', startangle=90)
axs[0, 0].set_title('Tienda 1')

# Gráfico para Tienda 2
axs[0,1].pie(vp2, labels=vp2.index, colors=['#11608f','#4f65a7', '#8a64b1', '#c15eaa', '#ed5b92', '#ff666e', '#ff8244', '#ffa600'],
             autopct='%1.1f%%', startangle=90)
axs[0,1].set_title('Tienda 2')

# Gráfico para Tienda 3
axs[1,0].pie(vp3, labels=vp3.index, colors=['#11608f','#4f65a7', '#8a64b1', '#c15eaa', '#ed5b92', '#ff666e', '#ff8244', '#ffa600'],
             autopct='%1.1f%%', startangle=90)
axs[1,0].set_title('Tienda 3')

# Gráfico para Tienda 4
axs[1,1].pie(vp4, labels=vp4.index, colors=['#11608f','#4f65a7', '#8a64b1', '#c15eaa', '#ed5b92', '#ff666e', '#ff8244', '#ffa600'],
             autopct='%1.1f%%', startangle=90)
axs[1,1].set_title('Tienda 4')

plt.show()
```

# 3. Calificación promedio de la tienda


```python
'''En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. 
El objetivo es conocer la satisfacción del cliente con los productos vendidos.'''
calif_promedio_tienda1 = tienda['Calificación'].mean()
calif_promedio_tienda2 = tienda2['Calificación'].mean()
calif_promedio_tienda3 = tienda3['Calificación'].mean()
calif_promedio_tienda4 = tienda4['Calificación'].mean()
x = ['Tienda 1','Tienda 2', 'Tienda 3', 'Tienda 4']
y = [calif_promedio_tienda1, calif_promedio_tienda2, calif_promedio_tienda3, calif_promedio_tienda4]

plt.plot(x,y, marker='o', color='#4f65a7', linestyle='-', linewidth=2, markersize=8)
plt.title('Grafico 3: Calificaciones promedio de los clientes por tienda')
```

# 4. Productos más y menos vendidos

```python
'''En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. 
Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.'''
# Graficar con subplots: 2 filas y 2 columnas
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.flatten()

# Lista de tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]
titulos = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, tienda in enumerate(tiendas):
    # Obtener el producto más y menos vendido
    mayor_venta = tienda.groupby('Producto')['Precio'].sum().idxmax()
    menor_venta = tienda.groupby('Producto')['Precio'].sum().idxmin()
    ventas_mas = tienda.groupby('Producto')['Precio'].sum().max()
    ventas_menos = tienda.groupby('Producto')['Precio'].sum().min()
    
    # Graficar barras para cada tienda
    axs[i].bar([mayor_venta, menor_venta], [ventas_mas, ventas_menos], 
               color=['mediumseagreen', 'salmon'], edgecolor='black')
    axs[i].set_title(titulos[i])
    axs[i].set_ylabel('Total vendido')
    axs[i].set_yscale('log')  # Escala logarítmica para ver mejor las diferencias

    # Mostrar los valores encima de las barras
    axs[i].text(0, ventas_mas, f'{ventas_mas:.2f}', ha='center', va='bottom', fontsize=8)
    axs[i].text(1, ventas_menos, f'{ventas_menos:.2f}', ha='center', va='bottom', fontsize=8)

plt.suptitle('Grafico 4: Productos más y menos vendidos', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Ajuste para que no se solapen los títulos
plt.show()


```

# 5. Envío promedio por tienda

```python
'''En este paso, debes calcular el costo de envío promedio para cada tienda. 
El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.'''
costo_envio = tienda['Costo de envío'].mean()
costo_envio2 = tienda2['Costo de envío'].mean()
costo_envio3 = tienda3['Costo de envío'].mean()
costo_envio4 = tienda4['Costo de envío'].mean()

plt.barh(['Tienda 1','Tienda 2', 'Tienda 3', 'Tienda 4'],[costo_envio, costo_envio2, costo_envio3, costo_envio4], 
        color=['#11608f', '#9d63b1', '#ff607b', '#ffa600'], 
        edgecolor='black'), plt.title('Grafico 5: Coste de envío promedio para cada tienda')
plt.show()
```

## 1. Introducción

El presente informe tiene como objetivo analizar el desempeño comparativo de cuatro tiendas del Challenge AluraStore para fundamentar una recomendación sobre cuál de ellas debería ser considerada para la venta. El análisis se basa en datos de ventas, costos y satisfacción del cliente extraídos de los archivos. Se evaluaron métricas clave como la facturación total, las ventas por categoría de producto, la calificación promedio otorgada por los clientes, el rendimiento de productos específicos (más y menos vendidos en términos de ingresos) y el costo promedio de envío. Las visualizaciones generadas (Grafico 1 a Grafico 5) respaldan las conclusiones presentadas.

## 2. Desarrollo: Análisis y Presentación de Datos

A continuación, se detallan los hallazgos para cada métrica analizada en las cuatro tiendas:


---


### 2.1. Análisis de Facturación (Ingresos Totales)

Se calculó la suma total de la columna 'Precio' para cada tienda, representando los ingresos brutos generados.

El **Grafico 1 (Gráfico de Barras)** muestra una comparación directa de estos ingresos. Se observa que la **Tienda 2** generó la mayor facturación, indicando una fortaleza en su capacidad de venta. Por el contrario,  la **Tienda 1 y la Tienda 4** presentaron la facturación más baja, lo que sugiere una debilidad potencial en su desempeño de ventas o posicionamiento en el mercado. La **Tienda 3** se ubico en una posiciones intermedia.

**Interpretación:** Una baja facturación puede ser un indicador clave para considerar la venta de una tienda, mientras que una alta facturación es una fortaleza significativa.



### 2.2. Ventas por Categoría

Se analizó la distribución de las ventas por 'Categoría del Producto' en cada tienda, contando el número de productos vendidos en cada categoría.

El **Grafico 2 (Gráficos Circulares)** ilustra la composición de las ventas por categoría para cada tienda. Por ejemplo, la **Tienda 3** mostró una fuerte concentración en **Muebles**, mientras que la **Tienda 2** tuvo una distribución más equilibrada entre sus categorias.

**Interpretación:** Una alta dependencia de categorías con baja demanda o margen podría ser una debilidad. Una buena diversificación o fortaleza en categorías populares es un punto a favor. La tienda con la peor mezcla o menor volumen en categorías clave podría ser candidata a la venta.



### 2.3. Calificación Promedio de la Tienda

Se calculó la media de la columna 'Calificación' para cada tienda, reflejando la satisfacción general del cliente.

El **Grafico 3 (Gráfico de Líneas)** compara estas calificaciones promedio. La **Tienda 3** obtuvo la calificación promedio más alta, lo que indica una alta satisfacción del cliente, una fortaleza importante. En contraste, la **Tienda 1 y la Tienda 4** recibieron la calificación más baja, sugiriendo problemas en la calidad del producto, el servicio o la experiencia de compra, lo cual representa una debilidad considerable.
Sin embargo algo a considerar es que no se diferencian demasiado las calificaciones promedio.

**Interpretación:** Una baja calificación promedio es un fuerte indicador de problemas operativos o de producto y una razón potencial para considerar la venta.



### 2.4. Productos Más y Menos Vendidos (por Ingresos)

Se identificó el producto que generó mayores ingresos ('Precio' total) y el que generó menores ingresos para cada tienda.

El **Grafico 4 (Gráficos de Barras)** muestra los ingresos generados por estos dos productos en cada tienda. Esto permite visualizar la magnitud de los productos estrella y aquellos con menor rendimiento en términos monetarios. Se puede notar si hay una gran disparidad entre el producto más y menos vendido en todas las tiendas. La menor disparidad se puede notar entre las **Tienda 1 y Tienda 4**

**Interpretación:** Una gran diferencia podría indicar una fuerte dependencia de pocos productos exitosos o una larga cola de productos de bajo rendimiento. Si el producto "más vendido" de una tienda genera significativamente menos ingresos que los de otras tiendas, podría ser una debilidad. 



### 2.5. Envío Promedio por Tienda

Se calculó el costo medio de la columna 'Costo de envío' para cada tienda, como un indicador de eficiencia logística o política de precios de envío.

El **Grafico 5 (Gráfico de Barras Horizontales)** compara estos costos promedio. Las **Tienda 1 y Tienda 4** presentó el costo de envío promedio más bajo, lo cual puede ser una ventaja competitiva o indicar eficiencia. La **Tienda 2** tuvo el promedio más alto, lo que podría impactar negativamente la rentabilidad o la percepción del cliente, representando una debilidad.

**Interpretación:** Un costo de envío promedio elevado puede ser un factor negativo a considerar al evaluar la viabilidad o atractivo de una tienda.
